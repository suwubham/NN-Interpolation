```python
def resize_nearest_neighbor(orig_img, new_width, new_height):
    #Create a new empty image with required dimensions.
    new_img = Image.new('RGB', (new_width, new_height)) 

    #Calcuate the scale factor for the axes
    x_scale = orig_img.width / new_width
    y_scale = orig_img.height / new_height

   #Iterate over each pixel in new image and find corresponding pixel in old image
    for x in range(new_width):
        for y in range(new_height):
            px = int(x * x_scale)
            py = int(y * y_scale)
            new_img.putpixel((x, y), orig_img.getpixel((px, py)))

    return new_img
```

```python
interpolated_image = resize_nearest_neighbor(original_image, 100, 100)
interpolated_array = np.array(interpolated_image)
interpolated_array.shape
```
